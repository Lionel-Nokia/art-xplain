17/03 : Création du dataframe à utiliser pour Streamlit

Constat : pour la v1 de la plateforme de sortie, on donne 3 suggestions de tableaux "similaires" et leur courant artistique. Ces tableaux sont évidemment compris dans notre database.
Pour apporter de la valeur ajoutée à l'utilisateur, il serait plus intéressant de donner le nom de l'artiste, le nom de son oeuvre, et la date.
Dans une V2, on pourra ensuite enrichir ce contenu avec un texte descriptif des oeuvres suggérées.

Point technique : on peut recréer un nouveau dataframe en utilisant le file-name composé de la manière suivante

/prénom-nom_titre_année

En utilisant la bibliothèque os, on peut utiliser toute la logique des chemins avec python. Dans data --> in -> courant artistique (colonne 5 du nouveau dataframe).
Il faudra évidemment mettre en page les chaînes de caractères extraites (majuscules pour commencer les noms et prénoms)

```python
import os
import pandas as pd

# 1. Définition du chemin racine (adapté selon le projet)
root_path = "data/in/kaggle-wikiart"
data_list = []

# 2. On parcourt l'arborescence
for root, dirs, files in os.walk(root_path):
    for file in files:
        # On ignore les fichiers systèmes ou parasites
        if not file.endswith(".jpg"):
            continue
            
        # Extraction du courant artistique (nom du dossier parent)
        courant = os.path.basename(root)
        
        # On enlève l'extension .jpg pour travailler sur le nom
        clean_name = file.replace(".jpg", "")
        
        # Logique de split : on sépare par les tirets "-"
        # Note : Il faut vérifier si tes noms d'oeuvres contiennent des tirets
        parts = clean_name.split("-")
        
        # On initialise les variables par défaut
        artiste = "Inconnu"
        oeuvre = "Inconnue"
        date = None # Vide si pas de donnée
        
        if len(parts) >= 3:
            artiste = parts[0].replace("_", " ").title()
            date = parts[-1] # La date est souvent à la fin
            # L'oeuvre est ce qu'il y a entre l'artiste et la date
            oeuvre = " ".join(parts[1:-1]).replace("_", " ").capitalize()
        
        # 3. On stocke les infos dans un dictionnaire
        data_list.append({
            "artiste": artiste,
            "oeuvre": oeuvre,
            "date_creation": date,
            "courant_artistique": courant,
            "path_name": os.path.join(root, file) # Ton chemin complet
        })

# 4. Création du DataFrame
df_art = pd.DataFrame(data_list)
```
