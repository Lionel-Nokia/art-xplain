# cmd

## 1 - Init. project

- Create a project template

    ```bash
    $ packgenlite art-xplain
    # create art-xplain directory
    $ cd art-xplain
    ```
- Check local branch 

    ```bash
    $ git branch
    * master
    ```
- Check remote status

    ```bash
    $ git remote -v
    ```
- Create remote repo

    ```bash
    $ gh repo create
    # "Push an existing local repository to github.com"
    ```

- Check remote creation status

    ```bash
    $ git remote -v

    origin  git@github.com:Lionel-Nokia/lekuto2.git (fetch)
    origin  git@github.com:Lionel-Nokia/lekuto2.git (push)
    ```
- View github repo

    ```bash
    $ gh repo view --web 
    ```

- View commits

    ```bash
    $ git log 
    
    commit b9c2a334dfe59bcfa8ab8d531dd7e95e9458891e (HEAD -> master, tag: 0.4, origin/master)
    Author: natarianni <lionel.natarianni@nokia-bell-labs.com>

    # Or
    $ git lg

    ```

- Push local dev to the remote (origin)

    ```bash
    $ git push origin master 
    ```

# Notes

- Project repo (git)

  - Project creation
    - from scratch:
      ```bash
      cd ~/code/YOUR_GITHUB_USERNAME
      mkdir YOUR_PROJECT_NAME
      cd YOUR_PROJECT_NAME
      ```
    - `$ packgenlite <PROJ_NAME>`
      - install:
        ```bash 
        $ ip install git+ssh://git@github.com/krokrob/packgenlite.git
        ```
    - or using `$ git init` in a new folder:
      ```bash
      git init
      echo '# YOUR_PROJECT_NAME' >> README.md
      git add README.md
      git commit -m 'kickstart YOUR_PROJECT_NAME'
      ```
    - then we have:
      - a `.git`file
        - for tracking
      - branche(s) 
        - `$ git branch`
          - -> master by default

    - Create remote repo
      - `$ gh repo create`
        - "Push an existing local repository to github.com"
      - `$ git remote -v `
        - display origin git@github ...
      - `$ git push origin master`
    
    - Add collaborator
      - github -> settings
      
    - Add Jobs in // and dipatch
      - For each Jon add tasks
      - A task = a branch

    - Branch strategy:
      Branch -> code feature -> commit(s) -> finish -> pull request -> review (with reviewer)
